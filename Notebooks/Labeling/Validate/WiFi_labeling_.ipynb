```
WiFi["Label"]=-1
WiFi.head()
```

```
flag=0
for i in range(len(Label)-1):
  print(i,flag)
  for j in range(flag,len(WiFi)):
    print(j)
    if (WiFi["Epoch time[ms]"][j]>=int(Label["Epoch time[ms]"][i]) and WiFi["Epoch time[ms]"][j]<int(Label["Epoch time[ms]"][i+1])):
      WiFi.loc[j, 'Label']=Label["Label"][i]
      flag=j+1
      break
    elif WiFi["Epoch time[ms]"][j]>int(Label["Epoch time[ms]"][i+1]):
      break
```
