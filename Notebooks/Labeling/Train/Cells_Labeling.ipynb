```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
Cells=pd.read_csv('/content/drive/MyDrive/SHL 2021/Train/Cells/Cells_Final.csv',low_memory=False)
Cells["Label"]=-1
Cells.head()
```

```
Label=pd.read_csv("/content/drive/MyDrive/SHL 2021/Train/Label/Label.csv",low_memory=False)
Label.head()
```

```
flag=0
for i in range(len(Label)-1):
  print(i, flag)
  for j in range(flag,len(Cells)):
    print(j)
    if (Cells["Epoch time[ms]"][j]>=int(Label["Epoch time[ms]"][i]) and Cells["Epoch time[ms]"][j]<int(Label["Epoch time[ms]"][i+1])):
      Cells.loc[j, 'Label']=Label["Label"][i]
      flag=j
      break
    elif Cells["Epoch time[ms]"][j]>int(Label["Epoch time[ms]"][i+1]):
      break
    else:
      j=j+1
```

```
Final_Cells=Cells[Cells['Label']!=-1]
Final_Cells.to_csv("/content/drive/MyDrive/SHL 2021/Train/Cells/Cells with label.csv",index=False)
```
